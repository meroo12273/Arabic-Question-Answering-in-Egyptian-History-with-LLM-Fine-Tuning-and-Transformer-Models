# Loading Data

```python
from langchain_community.document_loaders import PyPDFLoader

book_path = r'data.pdf'

loader = PyPDFLoader(book_path)
```

```python
pages = loader.load_and_split()
```

```python
print(f"this book contains {len(pages)} pages")
print(pages[1].page_content)
```

```python
import re
# clean pages
for page in pages:

    content = page.page_content

    # Remove references like [<any number>]
    content = re.sub(r'\[\d+\]', '', content)

    # Remove footnotes
    content = content.replace("\n", "").replace("\t", "").replace("-", "").replace("[", "").replace("]", "")



    page.page_content = content
```

# Split Book

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=10_000_000,
    chunk_overlap=50_000,
    length_function=len,
    is_separator_regex=False,
)
```

```python
splitted_docs = splitter.transform_documents(pages)
```

```python
print(f"Length of the Documents {len(splitted_docs)}")
```

```python
splitted_docs[0]
```

# Convert from `Documents` into plain text

```python
texts = []
for doc in splitted_docs:
    texts.append(doc.page_content)
```

# Embedding model

```python
model_kwargs = {'device': 'cuda'}
encode_kwargs = {'normalize_embeddings': False}

embedding_model_id = "asafaya/bert-large-arabic"
# embedding_model_id = "meta-llama/Meta-Llama-3-8B"
# embedding_model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

hf = HuggingFaceEmbeddings(
    model_name = embedding_model_id,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

# Database Creation

```python
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(splitted_docs, hf)
```

# Create Retriever

```python
db.save_local("faiss_db")
```

```python
retriever = db.as_retriever()
```
